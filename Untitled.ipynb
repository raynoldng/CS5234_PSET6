# Stream Counting Experiments

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
import pandas as pd
```

```python
from src.main import Experiment
from src.counter import StreamCounterMedian, StreamCounterNeighourMedian
```

```python
def run_experiment(counter_type, filename, A, B, text=False):

    if counter_type == 1:
        c = StreamCounterMedian(A, B)
    elif counter_type == 2:
        B = B+1 if B%2 != 0 else B
        c = StreamCounterNeighourMedian(A, B)

    if text:
        numbers = [w.strip() for w in open('datasets/'+filename).readlines()][:100000]
    else:
        numbers = [int(x.strip()) for x in open('datasets/'+filename).readlines()]
    exp = Experiment(numbers, c)
    err, elapsed = exp.run()
    return err, elapsed
```

```python
def batch_experiment(counter_type, filename, repeats, text=False):
    data = []
    for A in range(1, 10+1):
        B = int(100/A)
        results = [run_experiment(counter_type, filename, A, B, text=text) for i in range(repeats)]
        error = sum(r[0] for r in results) / repeats
        elapsed = sum(r[1] for r in results) / repeats
        data.append([A, B, error, elapsed])
    return data
```

```python
exp1 = batch_experiment(1, 'random.1000.in', 10)
exp2 = batch_experiment(2, 'random.1000.in', 10)
```

```python
exp3 = batch_experiment(1, 'exp.5.in', 10)
exp4 = batch_experiment(2, 'exp.5.in', 10)
```

```python
exp5 = batch_experiment(1, 'moby_dick.txt', 10, text=True)
exp6 = batch_experiment(2, 'moby_dick.txt', 10, text=True)
```

```python
exp7 = batch_experiment(1, 'exp.25.in', 10)
exp8 = batch_experiment(2, 'exp.25.in', 10)
```

```python
exp9 = batch_experiment(1, 'exp.125.in', 10)
exp10 = batch_experiment(2, 'exp.125.in', 10)
```

```python
columns = ['A', 'B', 'Error', 'Time (s)']

df1 = pd.DataFrame(exp1, columns=columns)
df2 = pd.DataFrame(exp2, columns=columns)

df3 = pd.DataFrame(exp3, columns=columns)
df4 = pd.DataFrame(exp4, columns=columns)

df5 = pd.DataFrame(exp5, columns=columns)
df6 = pd.DataFrame(exp6, columns=columns)

df7 = pd.DataFrame(exp7, columns=columns)
df8 = pd.DataFrame(exp8, columns=columns)

df9 = pd.DataFrame(exp9, columns=columns)
df10 = pd.DataFrame(exp10, columns=columns)
```
